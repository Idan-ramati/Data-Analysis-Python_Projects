```python
# 2. Based on the data frame you've created, display a histogram of Meta scores.
df_meta_score_60_90['Meta_score'].hist(bins=31, edgecolor='black')
plt.title('Histogram of Meta Scores')
plt.xlabel('Meta Score')
plt.ylabel('Frequency')
plt.show()
```
